```python
# Importar librerías

import pandas as pd
import numpy as np
import csv
from bs4 import BeautifulSoup
from datetime import datetime
import time
from time import sleep
import math
from textblob import TextBlob

# Librerías de Selenium

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

# Translator

from googletrans import Translator, constants
translator = Translator()
```

```python
# Twitter

def get_n_tweets(n, search_str='Covid 19'):
    option = Options()
    option.headless = False
    driver = webdriver.Chrome(options=option)
    
    driver.get("http://twitter.com/search?q=%24" + search_str + "&src=typeahead_click&f=top")
    driver.maximize_window()

    response = []
    for x in range(math.ceil(n / 10)):
        print(x)
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        try:
            WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "div[class='css-1dbjc4n r-1iusvr4 r-16y2uox r-1777fci r-kzbkwu']"))
            )

            e_tweets = driver.find_elements(By.CSS_SELECTOR, "div[class='css-1dbjc4n r-1iusvr4 r-16y2uox r-1777fci r-kzbkwu']")
            #print(len(e_tweets))

            for e_tweet in e_tweets:
                counter = 0
                while True:
                    try:
                        e_fullname = e_tweet.find_element(By.CSS_SELECTOR, "div>span[class='css-901oao css-16my406 r-poiln3 r-bcqeeo r-qvutc0']")
                        e_tweet_text = e_tweet.find_element(By.CSS_SELECTOR, "div[class='css-901oao r-18jsvk2 r-37j5jr r-a023e6 r-16dba41 r-rjixqe r-bcqeeo r-bnwqim r-qvutc0']")
                        e_tweet_date = e_tweet.find_element(By.CSS_SELECTOR, "a>time")
                        response.append({'date': e_tweet_date.text,'by': e_fullname.text,
                                         'tweet': e_tweet_text.text,
                                         'score': TextBlob(translator.translate(e_tweet_text.text).text).sentiment.polarity}),
                                         
                        break
                    except:
                        if counter < 1:
                            counter += 1
                            print(f'Failed to scrape {e_tweets.index(e_tweet)}. Trying again in 5 seconds.')
                            sleep(5)
                            continue
                        else:
                            break
        finally:
            continue
    driver.quit()
    return response
```

```python
# Obteniendo las noticias

tag = "BTC"
tweets = get_n_tweets(500 ,tag)
```

```python
tweets_df = pd.DataFrame(tweets)
tweets_df
```
